## Adresowanie ramki danych

Pamitasz jeszcze sowniki Pythona? Tam wartoci pobierao si na podstawie par **klucz - warto**. Znajc **klucz** miae dostp do **wartoci** poprzez zapis `slownik[klucz]`. Ramki danych i szeregi dziaaj podobnie. Znajc nazw kolumny i indeksu, moesz atwo nawigowa po obiekcie.

Zacznijmy od wybierania pojedynczych kolumn.

```python
df.columns
```

```python
df['price']  # Adresowanie takie samo jak w sowniku, dostalimy ca kolumn wartoci z indeksami
```

```python
df[['price', 'host_name']]  # Moemy wybra kilka kolumn na raz
```

Indeksowa mona rwnie po wartociach indeksu, co szczeglnie przydatne jest dla dat albo indeksw oznaczajcych unikaln rzecz (np.: numer oznaczajcy numer transakcji). By to jednak zrobi, musimy ustawi ktr z kolumn jako indeks - nasza ramka posiada indeks z wartociami typu cakowitego, ale wydaje si nie mie adnego specjalnego zastosowania poza utrzymywaniem unikalnoci w danych. Moglibymy ten indeks podmieni sprawdzajc czy ktra z kolumn ma same unikalne rekordy, ale zostawimy to zadanie na pniej.

Teraz po prostu wybierzymy wiersz z indeksem 38893.

Mamy na to trzy sposoby.

1. Parametr `loc[]` ktry przyjmuje dwie zmienne: `loc[nazwa indeksu, nazwa kolumny`:

```python
row_index = 38893

df.loc[row_index, :]
```

```python
df.loc[row_index, ]
```

```python
df.loc[row_index, ['longitude', 'latitude']]
```

```python
df['host_id'][row_index]  # Dlaczego to zadziaao? pojedyncza kolumna to obiekt typu series, dlatego takie adresowanie zadziaa
```

```python
df.iloc[30]  # 30 rekord w bazie do wyboru
```

```python
# Wszystkie wiersze gdzie `price` jest wiksze ni 1000

df[df['price'] > 1000]
```
