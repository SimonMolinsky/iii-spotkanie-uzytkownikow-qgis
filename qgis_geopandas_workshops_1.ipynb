Najważniejsze elementy okna:

- **nazwa notatnika**: w lewym, górnym rogu, przy zielono-żółto-niebieskim trójkącie usług **Google**. Na start ustawiona jako `Untitled.ipynb`, ale możesz dowolnie ją zmieniać. Konwencję nazywania notatników wybierz sam, ale dobrze mieć je w jakiś sposób uszeregowane, przykładowo za pomocą dat i opisane tytułami, które jasno i krótko sugerują zawartość, np.: `YYYY-MM-DD Walidacja danych.ipynb`. Czasami przeprowadzasz <i>-naście</i> eksperymentów z niewielkimi zmianami, wtedy warto przemyśleć bardziej skomplikowany sposób nazywania notatników,
- **pole wywołania kodu / komórka notatnika**: w centrum okna, z trójkątem przypominającym przycisk _play_ znany z odtwarzaczy. Jest to **komórka kodu**. Wyróżniamy jeszcze **komórki tekstowe**, czyli takie, w których przedstawiony jest ten notatnik,
- **pola wyboru komórki**: po lewej, u góry, pod paskiem narzędzi, znajdują się dwa przyciski `+ Kod` i `+ Tekst`. Klikając w nie, dodajesz do notatnika kolejno komórkę z kodem i komórkę z tekstem. To samo możesz zrobić inaczej, wybierając na pasku narzędzi `Wstaw -> Komórka z kodem` albo `Wstaw -> Komórka z tekstem`. Komórki możesz też wybierać dzięki polom `+ Kod` i `+ Tekst` które pojawiają się na środku ekranu, jeśli umieścisz kursor między komórkami,
- **pole komentarza** - w prawym górnym rogu znajduje się pole komentarza,
- **pole udostępnienia** - także znajduje się w prawym, górnym rogu. Oba pola działają podobnie do analogicznych pól w pakiecie **Google Docs**.

### Co możemy zrobić z notatnikiem?

Klikając w `Plik` na pasku narzędzi będziemy mogli zrobić kilka ważnych rzeczy, a wśród nich:

- utworzyć nowy notatnik,
- zapisać aktualny notatnik (autozapis istnieje, ale zawsze dobrym pomysłem jest zapisanie pracy przed wyłączeniem notatnika albo przed uruchomieniem problematycznego fragmentu kodu),
- pobrać notatnik,
- i jedno z najciekawszych poleceń: utworzyć kopię notatnika. Jak już pisałem wcześniej, czasami posiada się wiele kopii notatników w celu testów różnych hiperparametrów, którymi sterujemy modelami uczenia maszynowego.

Interesujące fakty na temat takiej komórki:

- jeśli chcemy uzyskać podgląd do zmiennej, nie musimy pisać `print(nazwa_zmiennej)`, wystarczy, że podamy nazwę zmiennej w ostatnim wierszu danej komórki. W naszym przykładzie inicjujemy zmienną `hello_world` zdaniem _Hello Google Colab_ a następniemy wpisujemy nazwę tej zmiennej i notatnik w szarej linii poniżej wyświetla nam jej zawartość. To bardzo przydatna cecha, raz, że przyspiesza eksperymentowanie, dwa, że pozwala zajrzeć do danych,
- żeby wywołać komórkę możesz kliknąć na przycisku `play` ale bardziej dogodną metodą jest użycie skrótu klawiaturowego. `Shift + Return (Enter)` to wywołanie kodu w aktualnej komórce i przejście do następnej a `Alt + Return (Enter)` to wywołanie kodu w aktualnej komórce i stworzenie następnej. Musisz uważać na tę drobną różnicę, bo możesz przypadkowo usunąć informacje wyjściowe!
- po prawej stronie od komórki, w której znajduje się kursor, położonych jest kilka ikonek narzędziowych: _strzałki_ pozwalają na przesuwanie komórki w górę i w dół wobec innych komórek w notatniku, _link do komórki_ pozwala ustawić hiperlink do wybranego fragmentu eksperymentu, _komentarz_ umożliwia komentowanie pojedynczych komórek, _usunięcie_ niszczy komórkę i jej zawartość (bezpowrotnie). Dobra rada: zanim zniszczysz komórkę, zastanów się dwa razy, czy nie lepiej przesunąć ją gdzieś na spód z odpowiednim komentarzem.

## Adresowanie ramki danych

Pamiętasz jeszcze słowniki Pythona? Tam wartości pobierało się na podstawie par **klucz - wartość**. Znając **klucz** miałeś dostęp do **wartości** poprzez zapis `slownik[klucz]`. Ramki danych i szeregi działają podobnie. Znając nazwę kolumny i indeksu, możesz łatwo nawigować po obiekcie.

Zacznijmy od wybierania pojedynczych kolumn.

```python
df.columns
```

```python
df['price']  # Adresowanie takie samo jak w słowniku, dostaliśmy całą kolumnę wartości z indeksami
```

```python
df[['price', 'host_name']]  # Możemy wybrać kilka kolumn na raz
```

Indeksować można również po wartościach indeksu, co szczególnie przydatne jest dla dat albo indeksów oznaczających unikalną rzecz (np.: numer oznaczający numer transakcji). By to jednak zrobić, musimy ustawić którąś z kolumn jako indeks - nasza ramka posiada indeks z wartościami typu całkowitego, ale wydaje się nie mieć żadnego specjalnego zastosowania poza utrzymywaniem unikalności w danych. Moglibyśmy ten indeks podmienić sprawdzając czy któraś z kolumn ma same unikalne rekordy, ale zostawimy to zadanie na później.

Teraz po prostu wybierzymy wiersz z indeksem 38893.

Mamy na to trzy sposoby.

1. Parametr `loc[]` który przyjmuje dwie zmienne: `loc[nazwa indeksu, nazwa kolumny`:

```python
row_index = 38893

df.loc[row_index, :]
```

```python
df.loc[row_index, ]
```

```python
df.loc[row_index, ['longitude', 'latitude']]
```

```python
df['host_id'][row_index]  # Dlaczego to zadziałało? pojedyncza kolumna to obiekt typu series, dlatego takie adresowanie zadziała
```

```python
df.iloc[30]  # 30 rekord w bazie do wyboru
```

```python
# Wszystkie wiersze gdzie `price` jest większe niż 1000

df[df['price'] > 1000]
```
